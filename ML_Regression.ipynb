A simple comparison of Actual Values and Predictions

```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.columns
```

```python
train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X_train = train.drop("SalePrice", axis = 1)
y_train = train.loc[:,"SalePrice"]
```

```python
model.fit(X_train, y_train)
```

```python
X_test = test.drop("SalePrice", axis = 1)
y_test = test.loc[:,"SalePrice"]
```

```python
# Now Let's predict
predictions = model.predict(X_test)
```

```python
y_test.shape #there should be only y values so 1 dimensional matrix
```

```python
predictions.shape #there should be only y^ values so 1 dimensional matrix
```

```python
comparison = pd.DataFrame({"Actual Values" : y_test, "Predictions" : predictions})
```

```python
comparison.head()
```

```python
comparison.tail()
```

Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
#Root Mean Squared Error:
rmse = sqrt(mean_squared_error(y_test, predictions))
rmse
```
